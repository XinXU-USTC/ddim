```python
import torch
import torchvision as tv
import torchvision.utils as tvu
import matplotlib.pyplot as plt
import numpy as np
dir = "/home/mathxuxin/ddim/exp/image_samples/"
cifar_dir = dir + "cifar10"
torch.manual_seed(1234)
```

# Sample Consistency in DDIMs

```python
nimages = 50000
idx = torch.randint(high = nimages, size = (20, 1))
idx = idx.squeeze(-1)
#idx = torch.tensor(range(200, 10))
dir_cifar_0 = cifar_dir + "-0-"
#steps = [10, 20, 50, 100, 1000]
steps = [10, 20]
l = []
for step in steps:
    for i in idx:
        l.append(tv.io.read_image(dir_cifar_0+str(step)+"/"+str(i.numpy())+".png"))
images = torch.stack(l)
img_grid = tvu.make_grid(images, nrow=20)
plt.figure(figsize=(20,20))
plt.ylabel("sample timesteps")
plt.yticks([18, 50, 85, 120, 155],[10, 20, 50, 100, 1000])
ax = plt.axes()
ax.xaxis.set_major_locator(plt.NullLocator())
plt.imshow(img_grid.permute(1, 2, 0))
```

```python
idx = torch.tensor(range(54, 74, 1))
dir_cifar_0 = cifar_dir + "-0-"
steps = [10, 20, 50, 100, 1000]
#steps = [10, 20]
l = []
for step in steps:
    for i in idx:
        l.append(tv.io.read_image(dir_cifar_0+str(step)+"/"+str(i.numpy())+".png"))
images = torch.stack(l)
img_grid = tvu.make_grid(images, nrow=20)
plt.figure(figsize=(20,20))
plt.ylabel("sample timesteps")
plt.yticks([18, 50, 85, 120, 155],[10, 20, 50, 100, 1000])
ax = plt.axes()
ax.xaxis.set_major_locator(plt.NullLocator())
plt.imshow(img_grid.permute(1, 2, 0))
```

# Sample Quality

```python
step = 10
noise = ["-0-", "-02-", "-05-", "-1-", "-ddpm-"]
idx = torch.randint(high = 40, size = (5, 1)).squeeze(-1)
l = []
for s in noise:
    for i in idx:
        aaa = cifar_dir + s + str(step) +"/" + str(i.numpy()) +".png"
        l.append(tv.io.read_image(aaa))
images = torch.stack(l)
img_grid = tvu.make_grid(images, nrow=5)
plt.figure(figsize=(10, 10))
plt.ylabel(r'$\eta$')
plt.yticks([18, 50, 85, 120, 155],[r'0.0', r'0.2', r'0.5', r'1.0', r'$\hat{\sigma}$'])
plt.xlabel(r'$dim(\tau)=10$')
ax = plt.axes()
ax.xaxis.set_major_locator(plt.NullLocator())
plt.imshow(img_grid.permute(1, 2, 0))
```
